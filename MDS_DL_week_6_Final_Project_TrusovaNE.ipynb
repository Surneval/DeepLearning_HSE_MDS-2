# Part 1: computer vision

## 1. Data analysis 

```python
import os
import json
import random
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Dataset, Subset
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
from sklearn.metrics import accuracy_score
from tqdm import tqdm


# Check for GPU support

if torch.cuda.is_available():  
    dev = "cuda:0" 
elif torch.backends.mps.is_built():
    dev = "mps"
else:
    dev = "cpu"  
    
device = torch.device(dev)
print(f"Using device: {device}")

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Load dataset paths
dataset_path = 'food-101'
train_ann_path = os.path.join(dataset_path, 'meta', 'train.json')
test_ann_path = os.path.join(dataset_path, 'meta', 'test.json')

# Load train and test annotations
with open(train_ann_path, 'r') as f:
    train_annotations = json.load(f)
with open(test_ann_path, 'r') as f:
    test_annotations = json.load(f)

# Set random seed for reproducibility
random.seed(42)

# Select a subset of classes (10 classes)
selected_classes = random.sample(list(train_annotations.keys()), 10)
train_annotations = {k: v for k, v in train_annotations.items() if k in selected_classes}
test_annotations = {k: v for k, v in test_annotations.items() if k in selected_classes}

# Visualize several examples from different classes
def visualize_examples(annotations, num_classes=5, num_examples=5):
    classes = random.sample(list(annotations.keys()), num_classes)
    fig, axes = plt.subplots(num_classes, num_examples, figsize=(15, 15))
    for i, cls in enumerate(classes):
        examples = random.sample(annotations[cls], num_examples)
        for j, example in enumerate(examples):
            img_path = os.path.join(dataset_path, 'images', f"{example}.jpg")
            img = plt.imread(img_path)
            axes[i, j].imshow(img)
            axes[i, j].axis('off')
            if j == 0:
                axes[i, j].set_title(cls, size=15)
    plt.show()

visualize_examples(train_annotations)

# Calculate basic statistics
num_classes = len(train_annotations)
num_examples = sum(len(v) for v in train_annotations.values())
class_distribution = Counter({cls: len(images) for cls, images in train_annotations.items()})
print(f'Number of classes: {num_classes}')
print(f'Number of examples: {num_examples}')
print('Class distribution:')
for cls, count in class_distribution.items():
    print(f'{cls}: {count}')

# Create a list of image paths and their labels
def create_image_list(annotations):
    image_list = []
    class_idx = {cls: idx for idx, cls in enumerate(annotations.keys())}
    for cls, images in annotations.items():
        for img in images:
            image_list.append((os.path.join(dataset_path, 'images', f"{img}.jpg"), class_idx[cls]))
    return image_list

train_images = create_image_list(train_annotations)
test_images = create_image_list(test_annotations)

# Custom dataset class
class CustomDataset(Dataset):
    def __init__(self, images, transform=None):
        self.images = images
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path, label = self.images[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label

train_dataset = CustomDataset(train_images, transform=transform)
test_dataset = CustomDataset(test_images, transform=transform)

# Split train dataset into train and validation sets
train_indices, val_indices = train_test_split(
    list(range(len(train_dataset))), test_size=0.1, random_state=42)

current_train_dataset = Subset(train_dataset, train_indices)
current_val_dataset = Subset(train_dataset, val_indices)

# Create data loaders with reduced batch size and single worker to avoid multiprocessing issues
batch_size = 16
train_loader = DataLoader(current_train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader = DataLoader(current_val_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
```

## 2. ResNet Fine-tuning

```python
# Load a pre-trained ResNet50 model
model = models.resnet50(pretrained=True)

# Modify the final layer to match the number of classes
num_classes = len(train_annotations)
model.fc = nn.Linear(model.fc.in_features, num_classes)

# Move the model to the appropriate device (MPS)
model = model.to(device)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Training loop with validation and progress tracking
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        print(f'Epoch {epoch + 1}/{num_epochs}')
        for inputs, labels in tqdm(train_loader, desc="Training"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)

        model.eval()
        val_loss = 0.0
        correct = 0
        with torch.no_grad():
            for inputs, labels in tqdm(val_loader, desc="Validation"):
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct += torch.sum(preds == labels.data)
        val_loss = val_loss / len(val_loader.dataset)
        val_acc = correct.float() / len(val_loader.dataset)

        print(f'Train Loss: {epoch_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}')
        
        # Clear cache to free up memory
        if torch.backends.mps.is_built():
            torch.mps.empty_cache()

    return model

# Evaluate the model on the test data
def evaluate_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Testing"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    accuracy = accuracy_score(all_labels, all_preds)
    return accuracy

if __name__ == '__main__':
    # Train the model
    trained_model = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10)
    
    # Evaluate the model on the test data
    test_accuracy = evaluate_model(trained_model, test_loader)
    print(f'Test Accuracy: {test_accuracy:.4f}')

```

```python
from transformers import ViTForImageClassification, ViTImageProcessor
from torch.optim import AdamW

# Load the image processor and model
image_processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224', num_labels=num_classes, ignore_mismatched_sizes=True)

# Move the model to the appropriate device (MPS)
model = model.to(device)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=1e-4)

# Transform function for ViT
def vit_transform(image):
    return image_processor(images=image, return_tensors="pt")["pixel_values"].squeeze(0)

# Custom dataset class for ViT
class ViTDataset(Dataset):
    def __init__(self, images, transform=None):
        self.images = images
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path, label = self.images[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label

# Create the train, val, and test datasets using the train and test splits created earlier
train_dataset_vit = ViTDataset([(train_images[i][0], train_images[i][1]) for i in train_indices], transform=vit_transform)
val_dataset_vit = ViTDataset([(train_images[i][0], train_images[i][1]) for i in val_indices], transform=vit_transform)
test_dataset_vit = ViTDataset(test_images, transform=vit_transform)

# Create data loaders with reduced batch size and single worker to avoid multiprocessing issues
train_loader_vit = DataLoader(train_dataset_vit, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader_vit = DataLoader(val_dataset_vit, batch_size=batch_size, shuffle=False, num_workers=0)
test_loader_vit = DataLoader(test_dataset_vit, batch_size=batch_size, shuffle=False, num_workers=0)

# Training loop with validation and progress tracking for ViT model
def train_vit_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        print(f'Epoch {epoch + 1}/{num_epochs}')
        for inputs, labels in tqdm(train_loader, desc="Training"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs).logits
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)

        model.eval()
        val_loss = 0.0
        correct = 0
        with torch.no_grad():
            for inputs, labels in tqdm(val_loader, desc="Validation"):
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs).logits
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct += torch.sum(preds == labels.data)
        val_loss = val_loss / len(val_loader.dataset)
        val_acc = correct.float() / len(val_loader.dataset)

        print(f'Train Loss: {epoch_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}')
        
        # Clear cache to free up memory
        if torch.backends.mps.is_built():
            torch.mps.empty_cache()

    return model

if __name__ == '__main__':
    # Train the ViT model
    trained_vit_model = train_vit_model(model, train_loader_vit, val_loader_vit, criterion, optimizer, num_epochs=10)
```

```python
def evaluate_vit_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Testing"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs).logits
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    accuracy = accuracy_score(all_labels, all_preds)
    return accuracy

# Evaluate the ViT model on the test data
test_accuracy_vit = evaluate_vit_model(trained_vit_model, test_loader_vit)
print(f'Test Accuracy of ViT Model: {test_accuracy_vit:.4f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, matthews_corrcoef
import torch
import numpy as np


# Task 4.1: Augment the training data with English examples from CoLA dataset
cola_df = pd.read_csv('in_domain_train.tsv', sep='\t', header=None, names=['id', 'labels', 'sentence'])
cola_df = cola_df[['sentence', 'labels']]

# Convert labels to integers
cola_df['labels'] = cola_df['labels'].apply(pd.to_numeric, errors='coerce')
cola_df = cola_df.dropna().astype({'labels': 'int'})

# Combine RuCoLa and CoLA datasets
train_df = pd.concat([train_df, cola_df])

# Convert labels to integers and handle non-integer values
train_df['labels'] = train_df['labels'].apply(pd.to_numeric, errors='coerce')
train_df = train_df.dropna().astype({'labels': 'int'})
val_df['labels'] = val_df['labels'].apply(pd.to_numeric, errors='coerce')
val_df = val_df.dropna().astype({'labels': 'int'})
dev_df['labels'] = dev_df['labels'].apply(pd.to_numeric, errors='coerce')
dev_df = dev_df.dropna().astype({'labels': 'int'})

# Task 1.2: Split the train dataset into train and validation sets
train_df, val_df = train_test_split(train_df, test_size=0.1, random_state=42)

# Prepare datasets for training
def preprocess_function(examples):
    return tokenizer(examples['sentence'], truncation=True, padding=True)

# Task 3.1: Use mBERT model
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

train_dataset = Dataset.from_pandas(train_df[['sentence', 'labels']])
val_dataset = Dataset.from_pandas(val_df[['sentence', 'labels']])
dev_dataset = Dataset.from_pandas(dev_df[['sentence', 'labels']])
datasets = DatasetDict({'train': train_dataset, 'val': val_dataset, 'dev': dev_dataset})

tokenized_datasets = datasets.map(preprocess_function, batched=True)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Define metrics
def compute_metrics(p):
    pred_labels = np.argmax(p.predictions, axis=1)
    acc = accuracy_score(p.label_ids, pred_labels)
    mcc = matthews_corrcoef(p.label_ids, pred_labels)
    return {'accuracy': acc, 'mcc': mcc}

# Task 4.2: Fine-tune mBERT on the augmented dataset
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=2)
model.to(device)

# Train and validate the model, and select the optimal number of epochs
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,  # Start with more epochs to select the best one
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['val'],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Train the model and evaluate after each epoch
eval_results = []
for epoch in range(1, training_args.num_train_epochs + 1):
    trainer.train(resume_from_checkpoint=None)
    eval_result = trainer.evaluate(eval_dataset=tokenized_datasets['val'])
    eval_results.append((epoch, eval_result['eval_accuracy'], eval_result['eval_mcc']))
    print(f"Epoch {epoch}: Accuracy = {eval_result['eval_accuracy']:.4f}, MCC = {eval_result['eval_mcc']:.4f}")

# Select the best epoch based on validation performance
best_epoch = max(eval_results, key=lambda x: x[1])[0]
print(f"Best epoch based on validation accuracy: {best_epoch}")

# Reconfigure the training arguments to use the best epoch
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=best_epoch,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)

# Reinitialize the trainer with the best number of epochs
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['val'],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Train the model again with the optimal number of epochs
trainer.train()

# Task 4.3: Evaluate the model on the development set
eval_result = trainer.evaluate(eval_dataset=tokenized_datasets['dev'])
print("Final evaluation on development set:", eval_result)

# Additional evaluation and analysis
eval_result_val = trainer.evaluate(eval_dataset=tokenized_datasets['val'])
print("Evaluation on validation set:", eval_result_val)

# Analyze the results and make conclusions
print(f"Accuracy on validation set: {eval_result_val['eval_accuracy']}")
print(f"MCC on validation set: {eval_result_val['eval_mcc']}")
print(f"Accuracy on development set: {eval_result['eval_accuracy']}")
print(f"MCC on development set: {eval_result['eval_mcc']}")
```
